## Modello di Ising

Il modello di Ising rappresenta un sistema di spin (questi possono assumere i valori discreti $\pm 1$) su un reticolo d-dimensionale con varie possibili condizioni al bordo. In questa esercitazione ci concentriamo su una catena monodimensionale di $N$ spin con condizioni periodiche al bordo.

Il sistema è pensato inizialmente per modellizzare i materiali ferromagnetici: in questa interpretazione degli spin up/down rappresenta la magnetizzazione totale del sistema. Più questi sono tra di loro allineati, maggiore è la magnetizzazione del sistema. Quando invece gli spin sono disordinati e non prevale nessun allineamento significa che le interazioni termiche hanno la meglio sulle interazioni tra gli spin. E' possibile matematizzare un modello di questo tipo con la seguente Hamiltoniana (assumiamo unità tali che $\mu_B=1$ and $k_B=1$):

$$
H = 
-J\sum_{i=1}^N s_i s_{i+1}
-\frac{h}{2}\sum_{i=1}^N (s_i + s_{i+1}) 
$$

dove $\{ s_i \}_{i=1,N}$ sono le variabili di spin uguali a $\{-1;1\}$, $J>0$ (modello ferromagnetico), e con condizioni periodiche al bordo $s_{N+1} = s_{1}$. Invece $h$ gioca il ruolo di un campo esterno.

A partire dalle particolari configurazioni di spin $\{s_i\}$ è possibile stimare le varie grandezze termodinamiche fondamentali:

1. $U(N,T) = \langle H \rangle_{h=0}$
2. $C(N,T) = k_B \beta^2 (\langle H^2 \rangle_{h=0}-\langle H \rangle_{h=0}^2)$ 
3. $\chi(N,T) = \beta \langle (\sum_{i=1,N}s_i)^2 \rangle_{h=0}$
4. $M(N,T,h=0.02)= \langle \sum_{i=1,N}s_i \rangle_{h=0.02}$

La probabilità che il sistema assuma una certa configurazione dipende, secondo la distribuzione di Boltzmann, dall'energia che questo assume:

$$p(\{s_i\})=\frac{\exp(-\beta E(\{s_i\}))}{Z}$$

E' allora possibile simulare il sistema campionando configurazioni in base alla loro probabilità di realizzarsi. Per fare questo nell'esercizio sono stati usati due diversi algoritmi di campionamento:

1. Metropolis sampling
2. Gibbs sampling

Entrambi esplorano lo spazio delle configurazioni capovolgendo spin e vedendo se l'energia del sistema aumenta o diminuisce. Per entrambi i tipi di campionamento la temperatura del sistema è mantenuta costante, definendo perciò il problema nell'insieme canonico. I due algoritmi sono stati implementati schematicamente nel seguente modo:

### Metropolis

1. Considera un configurazione $s=\{s_1,s_2,...,s_n\}$
2. Scegli in modo random uno degli $N$ spin e capovolgilo per ottenere la configurazione $s'=\{s_1,...,-s_i,...,s_n\}$
3. Accetta la nuova configurazione con una probabilità $A(s'|s)=\min\bigg[1,\frac{p(\{s_i\}')}{p(\{s_i\})}\bigg]$$=\min\bigg[1,\exp(-\beta\Delta E)\bigg]$, quindi accetta sempre quando il sistema va verso una configurazione a minore energia, mentre quelli a maggiore energia vengono accettati con una probabilità che diminuisce tanto più l'energia è alta.

### Gibbs

1. Considera un configurazione $s=\{s_1,s_2,...,s_n\}$
2. Scegli in modo random uno degli $N$ spin e ponilo uguale a +1 (porlo a -1 è una scelta del tutto equivalente) per ottenere la configurazione $s'=\{s_1,...,+1,...,s_n\}$
3. Fissa lo spin a +1 con una probabilità data da $\frac{1}{1+\exp(-\beta \Delta E)}$, altrimenti ponilo uguale a -1

Si vede quindi che l'algoritmo di Gibbs non rigetta mai alcun tentativo.

### Simulazione

L'obiettivo della simulazione è quello di generare varie configurazioni di spin, e con queste calcolare le grandezze termodinamiche precedentemente citate. Il fatto di conoscere la soluzione analitica del modello di Ising in una dimensione permette di confrontare i risultati ottenuti con la loro controparte analitica.

Per fare questo sono stati utilizzati entrambi gli algoritmi, i quali restituiscono configurazioni distribuite con peso di Boltzmann dopo un tempo adeguato di equilibrazione. 
Dal momento che la configurazione iniziale che viene generata è uno stato termico, ci si aspetta che per simulare il sistema a basse temperature sia necessario un tempo di equilibrazione maggiore rispetto che per sistemi a temperature più alte. Di seguito si riportano i grafici relativi all'equilibrazione di un sistema a diverse temperature, con campo magnetico esterno fissato a $h=0.02$ e $J=1$.

```python
import numpy as np
import matplotlib.pyplot as plt

points=100
T = np.linspace(0.2,3.0,num=points)
beta = 1/T
b = 1/T
h=0.02 
J = 1.0
Ns = 50
th = np.tanh(J/T)
thN= th**Ns
ch = 1/th

e = -J*( th + ch*thN )/( 1 + thN )
heat=((beta*J)**2)*(((1+thN+(Ns-1)*(th**2)+(Ns-1)*(ch**2)*thN)/(1+thN))-Ns*((th+ch*thN)/(1+thN))**2)

l1 = np.exp(b*J)*np.cosh(b*h)+np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
l2 = np.exp(b*J)*np.cosh(b*h)-np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
Z = l1**Ns + l2**Ns
M = (np.exp(b*J)*np.sinh(b*h)*((l1**(Ns-1))*(1+np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))) 
        + (l2**(Ns-1))*(1-np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J)))))/(Z)

X = beta*np.exp(2*beta*J)*(1-thN)/(1+thN)
```

Dall'andamento dell'energia possiamo vedere che, sia con Metropolis che con Gibbs, l'unica temperatura per cui sembra necessaria un'opportuna equilibrazione sia $T=0.5 K$. Negli altri casi la temperatura pare essere abbastanza alta affinchè la configurazione iniziale rappresenti bene già fin da subito i valori di energia attorno ai quali poi oscilla il sistema.

Per questo, per le simulazioni successive, si è pensato di abbondare e usare $10^3$ passi di equilibrazione.

### Simulazione

A questo punto facciamo la simulazione vera e propria, usando sempre sia Metropolis che Gibbs. Sono stati scelti unidici valori di temperatura compresi nell'intervallo $[0.5,2]$. Per ogni temperatura si è scelto di campionare $2* 10^6$ configurazioni e di divederle in $20$ blocchi. In questo modo è stato possibile calcolare con la media a blocchi (non mostriamo però l'andamento progressivo di tale media) il valore di aspettazione con la relativa incertezza per varie grandezze termodinamiche fondamentali, e confrontarle con il modello teorico.

Due ultime parole sull'implementazione di $M$, $C$ e $\chi$. Mentre la magnetizzazione è semplicemente data dalla media di tutti gli spin, il calore specifico e la suscettività sono calcolate come:

$$C=k_B\beta^2 \left(\langle H^2 \rangle -\langle H \rangle^2 \right)$$ e 
$$\chi=\beta \left[ \left\langle \left( \sum_{i=1,N}s_i \right)^2 \right\rangle - \left\langle \sum_{i=1,N}s_i \right\rangle^2 \right] $$

Nel codice si è quindi fatto in modo di conservare $H^2$ e $\langle \left( \sum_{i=1,N}s_i \right)^2 \rangle$ (le altre grandezze sono già salvate attraverso l'energia e la magnetizzazione) per poter così eseguire il calcolo.

```python
points=100
T = np.linspace(0.4,2.1,num=points)
b = 1/T
J = 1.0
Ns = 50
th = np.tanh(J/T)
thN= th**Ns
ch = 1/th
N=50

l1 = np.exp(b*J)*np.cosh(b*h)+np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
l2 = np.exp(b*J)*np.cosh(b*h)-np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
Z = l1**Ns + l2**Ns


EnEx = -J*( th + ch*thN )/( 1 + thN )
HeatEx=((b*J)**2)*(((1+thN+(Ns-1)*(th**2)+(Ns-1)*(ch**2)*thN)/(1+thN))-Ns*((th+ch*thN)/(1+thN))**2)
MagEx = (np.exp(b*J)*np.sinh(b*h)*((l1**(Ns-1))*(1+np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))) 
        + (l2**(Ns-1))*(1-np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J)))))/(Z)
XEx = b*np.exp(2*b*J)*(1-thN)/(1+thN)
```

```python
def UExact(T,J,N):
    beta = 1/T
    frac = (1+np.tanh(beta*J)**(N-2))/(1+np.tanh(beta*J)**N)
    return -J*np.tanh(beta*J)*frac

def ChiExact(T,J,N):
    beta = 1/T
    tanhN = np.tanh(beta*J)**N
    return beta*np.exp(2*beta*J)*(1-tanhN)/(1+tanhN)

def HeatExact(T,J,N):
    beta = 1/T
    th = np.tanh(J/T)
    thN= th**N
    ch = 1/th
    heat=((beta*J)**2)*(((1+thN+(N-1)*(th**2)+(N-1)*(ch**2)*thN)/(1+thN))-N*((th+ch*thN)/(1+thN))**2)
    return heat

def MagExact(T,J,N,h):
    b=1./T
    l1 = np.exp(b*J)*np.cosh(b*h)+np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
    l2 = np.exp(b*J)*np.cosh(b*h)-np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
    Z = l1**N + l2**N
    M = (np.exp(b*J)*np.sinh(b*h)*((l1**(N-1))*(1+np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))) 
        + (l2**(N-1))*(1-np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J)))))/(Z)
    return(M)

x  = np.linspace(0.4,2.1)
yU = UExact(x,1.,50.)
yX = ChiExact(x,1.,50.)
yH = HeatExact(x,1.,50.)
yM = MagExact(x,1.,50.,0.02)
zero = np.zeros(x.size)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Plot the first graph in the first subplot
ax1.errorbar(temp,energy,yerr=enerr, color='red',fmt='.')
ax1.plot(T, EnEx,'--',color='orange')
ax1.set_title('Metropolis Sampling')
ax1.set_xlabel('Temperatura')
ax1.set_ylabel('Energia')
ax1.grid(True)

# Plot the second graph in the second subplot
ax2.errorbar(temp2,energy2,yerr=enerr2, color='blue',fmt='.')
ax2.plot(T, EnEx,'--',color='orange')
ax2.set_title('Gibbs Sampling')
ax2.set_xlabel('Temperatura')
ax2.set_ylabel('Energia')
ax2.grid(True)

#compatibility test
plt.figure(figsize=(14,6))
plt.errorbar(temp,energy-UExact(temp,J,N),enerr,label='Metropolis algorithm',fmt='o',capsize=5, capthick=1.5, color='red')
plt.errorbar(temp2,energy2-UExact(temp2,J,N),enerr2,label='Gibbs algorithm',fmt='o',capsize=5, capthick=1.5, color='blue')
plt.plot(temp,np.zeros(temp.size),'--',color='orange')
plt.grid(True)
plt.xlabel('$T $',fontsize=14)
plt.ylabel('$U/N - U/N_{model}$',fontsize=14)
plt.legend()
```
