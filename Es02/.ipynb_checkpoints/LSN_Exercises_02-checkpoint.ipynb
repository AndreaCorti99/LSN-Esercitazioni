### Exercise 02.1
- Compute the following 1D integral via Monte Carlo 
$$I = \int_0^1 \frac{\pi}{2}\cos(\pi x/2) dx = 1$$

1. sampling a uniform distribution in $[0,1]$
2. using importance sampling (i.e. sampling a non-uniform probability in $[0,1]$)

<span style="color:red">Show a picture of both your estimations of $I$</span> and their uncertainties with a large number of *throws* $M$ (e.g. $M\ge 10^4$) as a function of the number of blocks, $N$</font>.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
def error(AV,AV2,n):  # Function for statistical uncertainty estimation
    if n==0:
        return 0
    else:
        return math.sqrt((AV2[n] - AV[n]**2)/n)

def integrand(x):
    return np.pi/2*np.cos(np.pi*x/2)

M=10**6
N=100
L=int(M/N)            # Number of throws in each block, please use for M a multiple of N
np.random.seed(1)     # Fixing random seed for reproducibility
r = np.random.rand(M) # U[0,1) uniform distribution
        
x = np.arange(N)      # [0,1,2,...,N-1]
ave = np.zeros(N)
av2 = np.zeros(N) #tiene i quadrati delle medie
sum_prog = np.zeros(N)
su2_prog = np.zeros(N)
err_prog = np.zeros(N)

for i in range(N):
    sum1 = 0
    for j in range(L):
        k = j+i*L
        sum1 += integrand(r[k])
    ave[i] = sum1/L     #media sui numeri appartenenti al blocco i-esimo
    av2[i] = (ave[i])**2 # (r_i)^2 

for i in range(N):
    for j in range(i+1):
        sum_prog[i] += ave[j] # SUM_{j=0,i} r_j
        su2_prog[i] += av2[j] # SUM_{j=0,i} (r_j)^2
    sum_prog[i]/=(i+1) # Cumulative average
    su2_prog[i]/=(i+1) # Cumulative square average
    err_prog[i] = error(sum_prog,su2_prog,i) # Statistical uncertainty
    
plt.errorbar(x,sum_prog,yerr=err_prog)
plt.xlabel('#throws')
plt.ylabel('<r>-1/2')
plt.grid(True)
plt.show()
```

```python
#ATTENZIONE FAR ANDARE LA CELLA PRECEDENTE A QUESTA OGNI VOLTA

def function(x):
    return np.pi*np.cos(np.pi*x/2)/(4*(1-x))

def prob(x):
    return 2*(1-x)

def LinDist(N):  # Generate exponential distributed numbers
    un = np.random.rand(N)
    LinD = 1-np.sqrt(1-un)
    return LinD

np.random.seed(1)     # Fixing random seed for reproducibility
s = LinDist(M) # U[0,1) uniform distribution
        
#salvo i dati dell'integrale calcolato in modo uniforme in due vettori
#CAMBIARE NOMI SUM PROG ERR PROG IN MODO DA NON DOVER FAR ANDARE OGNI VOLTA ENTRAMBE
#LE CELLE
sum_progUn = sum_prog
err_progUn = err_prog
    
x = np.arange(N)      # [0,1,2,...,N-1]
ave = np.zeros(N)
av2 = np.zeros(N) #tiene i quadrati delle medie
sum_prog = np.zeros(N)
su2_prog = np.zeros(N)
err_prog = np.zeros(N)

for i in range(N):
    sum1 = 0
    for j in range(L):
        k = j+i*L
        sum1 += function(s[k])
    ave[i] = sum1/L     #media sui numeri appartenenti al blocco i-esimo
    av2[i] = (ave[i])**2 # (r_i)^2 

for i in range(N):
    for j in range(i+1):
        sum_prog[i] += ave[j] # SUM_{j=0,i} r_j
        su2_prog[i] += av2[j] # SUM_{j=0,i} (r_j)^2
    sum_prog[i]/=(i+1) # Cumulative average
    su2_prog[i]/=(i+1) # Cumulative square average
    err_prog[i] = error(sum_prog,su2_prog,i) # Statistical uncertainty

plt.errorbar(x,sum_progUn,yerr=err_progUn,alpha=0.6)
plt.errorbar(x,sum_prog,yerr=err_prog,alpha=0.8)
plt.xlabel('#throws')
plt.ylabel('ordinate')
plt.ylim([0.998,1.003])
plt.grid(True)
plt.show()
```

### Exercise 02.2
- <span style="color:red">3D Random Walks (RW) on a cubic lattice and in the continuum:</span> Repeat many times (e.g. say $10^4$) the simulation of a random walk in 3D always starting at the origin

1. on a cubic lattice with lattice constant $a=1$; at each discrete time the walker makes a forward or backward step of length equal to $a$ in one of the 3 principal directions of the lattice: $x$, $y$ or $z$
2. in the continuum; at each discrete time the walker makes a step of length equal to $a(=1)$ along a **random direction** obtained by sampling **uniformly** the solid angle: $\theta \in [0,\pi]$ and $\phi \in [0,2\pi]$

<span style="color:red">Show a picture of $\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$ for both RWs</span>, with their statistical uncertainties, as function of the step $i\in [0,10^2]$.

Suggestion: divide your $M$ throws into a reasonable number of blocks, compute $\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$ as function of the step $i\in [0,10^2]$ in each block and use these values (for fixed $i$) to obtain the average value and its statistical uncertainty.

Note that you could try to fit both results with a function like $f(N)=k\sqrt{N}$. Do your results indicate a diffusive behavior?

```python
from scipy.optimize import curve_fit

def f (x, k):
    return k * np.sqrt(x)

yD, errD=np.loadtxt("c++/outfile221.txt", unpack=True)
yC, errC=np.loadtxt("c++/outfile222.txt", unpack=True)

N=yD.shape[0]
x=np.arange(1,N+1)


#discrete fit
p_opt_disc, p_cov_disc = curve_fit(f, x, yD)
y_fit = f(x,p_opt_disc[0])

print( ' --- Random Walk over a discrete lattice --- ')
print('fit with k*sqrt(x): k = ',p_opt_disc)
print('confidence over k : +- ',np.sqrt(p_cov_disc))

#continuous fit
p_opt_cont, p_cov_cont = curve_fit(f, x, yC)

print( ' --- Random Walk in a continuum space --- ')
print('fit with k*sqrt(x): k = '+str(p_opt_cont))
print('confidence over k : +- '+str(np.sqrt(p_cov_cont)))


#graphing
plt.figure(figsize=(10,10))

#discrete RW
plt.subplot(211)
plt.errorbar(x,yD,errD,fmt='.',label='MonteCarlo data', markersize=1.5, ecolor="black")
plt.grid(True)
plt.xlabel('steps',fontsize=13)
plt.ylabel(r'$\sqrt{\langle|\vec{r}_{N}|^2\rangle}$',fontsize=13)
plt.title("Discrete Lattice Random Walk",fontsize=16)
y_fitD = f(x,p_opt_disc)
plt.plot(x,y_fitD,'r',linewidth=1,label='$f(x)=k\sqrt{x}$')
plt.legend()

#continuous RW
plt.subplot(212)
plt.errorbar(x,yC,errC,fmt='.',label='MonteCarlo data', markersize=1.5, ecolor="black")
plt.grid(True)
plt.xlabel('steps',fontsize=13)
plt.ylabel(r'$\sqrt{\langle|\vec{r}_{N}|^2\rangle}$',fontsize=13)
plt.title("Continuum Random Walk",fontsize=16)
y_fitC = f(x,p_opt_cont)
plt.plot(x,y_fitC,'r',linewidth=1,label='$f(x)=k\sqrt{x}$')
plt.legend()
```
