## Exercise 07.1

In questo esercizio, a differenza del 4, useremo l'algoritmo fornito in modalità Monte Carlo. Questo significa considerare il sistema a temperatura fissata (NVT, ovvero ensemble canonico). Al contrario dell'esercizio 4 l'energia non sarà quindi fissata e il sistema si distribuirà secondo:

$$P=\frac{1}{Z}\exp(-\beta E)$$

L'algoritmo si occupa quindi di campionare questa distribuzione seguendo di volta in volta l'evoluzione dei parametri macroscopici.

### Correzioni di coda

Nel codice MD/MC è previsto un raggio di cut-off oltre il quale non vengono più calcolate le interazioni. E' possibile correggere questa approssimazione aggiungendo dei termini detti "tail corrections". Per quanto riguarda l'energia potenziale e la pressione in un sistema con potenziale alla Lennard-Jones si dimostra che i termini sono dati dalle seguenti espressioni:

$$\frac{\langle U \rangle _{tail}}{N}   = \frac{8\pi\rho\epsilon}{9 r_c^9} - \frac{8\pi\rho\epsilon}{3 r_c^3}$$

e

$$\frac{\langle W \rangle _{tail}}{3N\epsilon} = 32\pi\rho \left[\frac{1}{9 r_c^9} -\frac{1}{6 r_c^3}\right]$$

L'obiettivo dell'esercizio è calcolare energia e pressione tenendo conto di queste correzioni.

## Autocorrelazione

Dal momento che l'algoritmo di Metropolis sceglie la nuova posizione in base a un salto che parte da quella vecchia, configurazioni successive sono evidentemente correlate. Diventa quindi interessante capire quanti passi Monte Carlo sono necessari affinchè grandezze contigue risultino scorrelate. Una volta definito il tempo di correlazione $\tau$ come costante di decadimento dell'esponenziale che ci aspettiamo descrivere la correlazione, possiamo fare il data-blocking considerando blocchi con un numero di elementi molto maggiore di $\tau$. In questo modo non si sottostima l'incertezza associata ai blocchi.

L'autocorrelazione viene calcolata usando la seguente formula:

$$\chi(t) = \frac{\frac{1}{t_{max}-t}\sum_0^{t_{max}-t} m(t')m(t'+t)-\frac{1}{t_{max}-t}
              \sum_0^{t_{max}-t} m(t')\times\frac{1}{t_{max}-t}\sum_0^{t_{max}-t} m(t'+t)}
              {\frac{1}{t_{max}}\sum_0^{t_{max}} m^2(t')-\left(\frac{1}{t_{max}}\sum_0^{t_{max}} m(t')\right)^2 }
$$          

```python
N = 100000
tmax = N
taumax = 200

colors=['red','blue','green']

plt.figure(figsize=(8,6))

for i in range(0,3):

    chi = []

    print("\ntau (nth of "+str(taumax)+") - "+phase[i]+": ")
    for tau in range(taumax):
        s1=0
        s2=0
        s3=0
        s4=0
        s5=0

        # autocorrelation: see lect.6, p.31
        for tp in range(tmax-tau):
            s1 += Epot[i][tp]*Epot[i][tp+tau]
            s2 += Epot[i][tp]
            s3 += Epot[i][tp+tau]

        for tp in range(tmax):
            s4 += Epot[i][tp]*Epot[i][tp]
            s5 += Epot[i][tp]

        fatt = 1/(tmax-tau)
        fatt2 = 1/tmax

        num = fatt*s1-fatt*s2*fatt*s3
        den = fatt2*s4 - (fatt2*s5)*(fatt2*s5)

        chi.append(num/den)

        if tau%10==0:
            print(tau, end= ", ")

    plt.plot(chi, color = colors[i], label = phase[i])
    plt.xlabel("Monte carlo time (~ distance between different measurments)")
    plt.ylabel("Autocorrelation")
    plt.legend()
    plt.grid()

plt.title("Autocorrelation vs time distance", fontsize = 18, fontweight = "bold")
plt.show
```

Come anticipato, se i blocchi non sono abbastanza grandi c'è correlazione e ci aspettiamo un'incertezza sottovalutata. Controlliamo esplicitamente questo fatto calcolando le incertezze in funzione della grandezza del blocco. Consideriamo $L$ che va da $10$ fino a $5*10^3$ a passi di dieci e calcoliamo l'incertezza con la seguente formula:

$$\sigma_{\langle g \rangle}=\sqrt{\frac{1}{N-1}\left[\frac{1}{N}\sum_1^N g_i^2-\left(\sum_1^N g_i\right)^2\right]}$$

```python
# # # functions # # #
def Error(ave, ave2, n):
    if(n==0): 
        return 0
    return np.sqrt((ave2-ave**2)/n)

def dataBlocking(x, N, L):
    mean=0
    mean2=0
    error=0
    for n in range(N):
        mean_block=0
        for l in range(L):
            mean_block+=x[l+n*L]
        mean_block/=L
        mean=(mean*n+mean_block)/(n+1)
        mean2=(mean2*n+mean_block**2)/(n+1)
        error=Error(mean, mean2,n)
    return error
# ----------------------------------------------------#

L_m=10
L_M=5000
M=500000

fig=plt.figure(figsize=(17, 17))

for i in range(0,3):
    plt.subplot(3,1,i+1)
    plt.title("Statistical error - "+phase[i], size='xx-large', weight='bold')

    y=Epot[i] # data loaded in 7.1
    error=[]
    x=[]

    L_range = range(10,5000, 10)

    print("\nL (of "+str(L_M)+") - "+phase[i]+": ")
    for L in L_range: 
        if L%500==0:
            print(L, end = ", ")
        N=int(M/L)
        err=dataBlocking(y,N,L)
        error.append(err)
        x.append(L)

    plt.plot(x, error, label = phase[i], color = colors[i])
        
    plt.ylabel("Error")
    plt.xlabel('L')
    #plt.xscale("log")
    plt.grid(True)
        
plt.show()
```

In effetti vediamo che per blocchi di piccola dimensione l'incertezza viene sottostimata. Aumentando la dimensione dei blocchi l'incertezza ha invece un andamento oscillante intorno a valori maggiori.

## Esercizio 07.3

In questa sezione, abbiamo implementato l'algoritmo per il calcolo della funzione di distribuzione radiale $g(r)$. Abbiamo suddiviso l'intervallo $[0,L/2]$ (dove $L$ rappresenta la dimensione di un lato del box) in 100 intervalli al fine di rappresentare la funzione tramite un istogramma. Durante l'esecuzione del programma, vengono esaminate tutte le coppie di particelle, e quando due particelle vengono individuate a una distanza compresa all'interno di un intervallo $[r,r+dr]$, viene incrementato di $+2$ il contatore corrispondente a quel determinato intervallo. Infine, l'istogramma viene normalizzato dividendo il contenuto di tutti i bin per un fattore $\rho N \Delta V$ con $\Delta V=4/3\pi((r+dr)^3-r^3)$, dove $N$ rappresenta il numero totale di particelle e $\rho$ la densità. Su questi dati viene sempre fatta la media a blocchi che permette una stima di $g(r)$ con la relativa incertezza.

Il calcolo viene fatto con 50 blocchi da 2000 passi l'uno.

```python
plt.figure(figsize=(15, 7))
x=np.arange(0,100)
phase=["gas", "liquid", "solid"]

# box scaling
rho = [0.05, 0.8, 1.1]
npart = 108
L = [0.0,0.0,0.0]
for i in range(3):
    L[i] = ((rho[i]/npart)**(-1/3))/200

for i in range (3):
        gdr = np.loadtxt("c++/output/"+phase[i]+"/7.3/output_G.dat", usecols=range(0,100), unpack='true', max_rows=20)
        y = np.mean(gdr, axis=1)
        err = np.std(gdr, axis=1)
        plt.errorbar(x*L[i], y, err, marker='',label=phase[i] )
        plt.title("Radial distribution function $g(r)$", fontsize=20)
        plt.xlabel('Normalized distance', fontsize=15)
        plt.ylabel('g(r)', fontsize=15)
        plt.legend()
        plt.grid(True)
        plt.xlim(0.5,2.5)
plt.show()
```

Nessuna particella si trova al di sotto della distanza in cui l'hard core repulsivo del potenziale diverge, il che rende il profilo della distribuzione compatibile con il potenziale di Lennard-Jones. Si osserva invece che tutte e tre le fasi presentano un guscio intorno al minimo del potenziale, con una maggiore chiarezza nella fase solida, come previsto. Oltre al guscio principale, sia la fase liquida che quella solida mostrano la presenza di ulteriori gusci, con i gusci della fase solida più definiti a indicare una struttura più regolare rispetto che le altre fasi.

### Exercise 07.4

Simuliamo quindi un sistema di Argon ($\sigma = 0.34$ nm, $\epsilon/k_B = 120$ K, $m=39.948$ amu) nelle solite condizioni:

1. Fase gassosa: $\rho^\star = 0.05$, $T^\star = 1.2$ (raggio di cut-off: $r_c = 5.0$)
2. Fase liquida: $\rho^\star = 0.8$, $T^\star = 1.1$ (raggio di cut-off: $r_c = 2.5$)
3. Fase solida: $\rho^\star = 1.1$, $T^\star = 0.8$ (raggio di cut-off: $r_c = 2.2$)

Mostriamo quindi la stima della funzione radiale fatta sia in ensemble NVT con simulazione Monte Carlo, sia in ensemble NVE con algoritmo di Verlais. Confrontiamo le distribuzioni per vedere se c'è accordo.

```python
rho = [0.05, 0.8, 1.1]
npart = 108
L = [0.0,0.0,0.0]
Ph=["Gas","Liquid","Solid"]

for i in range(3):
        L[i] = ((rho[i]/npart)**(-1/3))/200
    
x=np.arange(0,100)
    
for i in range (3):
        fig=plt.figure(figsize=(15, 5))
        
        plt.subplot(1,2,1)
        gdr = np.loadtxt("c++/output/"+phase[i]+"/7.3/output_G.dat", usecols=range(0,100), unpack='true', max_rows=50)
        gdr2 = np.loadtxt("c++/output/"+phase[i]+"/7.4/output_G.dat", usecols=range(0,100), unpack='true', max_rows=50)
        y = np.mean(gdr, axis=1)
        y2 = np.mean(gdr2, axis=1)
        err = np.std(gdr, axis=1)
        err2 = np.std(gdr2, axis=1)
        plt.errorbar(x*L[i], y, err, marker='', color='green',alpha=0.8, label='MC')
        plt.errorbar(x*L[i], y2, err2, marker='', color='purple',alpha=0.5,label='MD')
        plt.title('$g(r)$ (50blks, 2000step)', size='xx-large', weight='bold')
        plt.ylabel('$g(r)$')
        plt.xlabel('Normalized Distance')        
        plt.legend()
        plt.grid(True)
        
        plt.subplot(1,2,2)
        plt.errorbar(x*L[i], y-y2, err+err2, marker='', color='darkblue', ecolor='skyblue')
        plt.title('$g_{MC}(r)-g_{MD}(r)$ (50blks, 2000step)', size='xx-large', weight='bold')
        plt.ylabel('$g_{MC}(r)-g_{MD}(r)$')
        plt.xlabel('Normalized Distance')
        plt.grid(True)
        
        fig.text(-0.01, 0.5, Ph[i],  weight="bold", size=20)
        
        plt.show()
```

L'incertezza complessiva, per essere conservativi, è stata calcolata con una somma algebrica delle incertezze ottenute dai due diversi metodi. Notiamo quindi un ottimo accordo tra le distribuzioni, sempre all'interno di $1\sigma$. E' interessante notare che le due simulazioni agiscono in due ensemble termodinamici differenti. Tuttavia, grazie all'equilibrazione della temperatura nella simulazione NVE, i due sistemi vengono portati in condizioni termodinamiche che sono confrontabili, e che restituiscono quindi la stessa distribuzione radiale.
